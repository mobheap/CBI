```python
from pathlib import Path
import pandas as pd
```

```python
BASE_DIR = Path().resolve().parent
DATA_DIR = BASE_DIR / 'data'

tools = pd.read_csv(DATA_DIR / 'Main_Tools.csv')
features = pd.read_csv(DATA_DIR / 'Uptrace.csv')
```

```python
print(tools.columns)
print(features.columns)
```

```python
tools.Name
```

```python
names = [name for name in tools.Name]
```

```python
features = features[features['Tool'].isin(names)]
features
```

```python
features = features.rename(columns={'Tool': 'Name'})
```

```python
data = pd.merge(tools, features, how='outer', on='Name', suffixes=('',''))
data
```

```python
data.to_csv(DATA_DIR/ 'tools.csv')
```
