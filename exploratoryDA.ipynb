```python
import pandas as pd
```

```python
data_Gartner = pd.read_csv("APM tools Gartner.csv")
data_G2 = pd.read_csv("APM tools G2.csv")
```

```python
print(data_G2.columns, '\n', data_Gartner.columns)
```

```python
data_Gartner.drop('Description', axis=1, inplace=True)
```

```python
data = pd.merge(data_Gartner, data_G2, how='outer', on='Name', suffixes=('Gartner','G2'))
data
```

```python
data['Number of Reviews'] = data['Number of reviewsGartner'].fillna(0) + data['Number of reviewsG2'].fillna(0)
data.drop(columns=['Number of reviewsGartner', 'Number of reviewsG2'], inplace=True)
data
```

```python
data['Rating-5G2'].fillna(data['Rating-5Gartner'], inplace=True)
data['Rating-5Gartner'].fillna(data['Rating-5G2'], inplace=True)
data
```

```python
data['Rating'] = data[['Rating-5Gartner', 'Rating-5G2']].mean(axis=1)
data.drop(columns=['Rating-5Gartner', 'Rating-5G2'], inplace=True)
data = data[['Name', 'Rating', 'Number of Reviews']]
data
```

```python
data.to_csv('APM_tools.csv', index=False)
```
