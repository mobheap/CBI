```python
import pandas as pd
```

```python
data = pd.read_csv('RESULT.csv')
data
```

```python
mapping_openT = {
    "Limited": 0,
    "Yes": 1
}

mapping_ease = {
    "Easy" : 1,
    "Average": 0
}

mapping_yes = {
    "Yes": 1
}
```

```python
data['OpenTelemetry Support'] = data['OpenTelemetry Support'].map(mapping_openT)
data['Distributed Tracing'] = data['Distributed Tracing'].map(mapping_yes)
data['Ease of Use'] = data['Ease of Use'].map(mapping_ease)
data['Real User Monitoring (RUM)'] = data['Real User Monitoring (RUM)'].map(mapping_yes)
data['Log Management'] = data['Log Management'].map(mapping_yes)
data['Mobile Monitoring'] = data['Mobile Monitoring'].map(mapping_yes)
data
```

```python
cols_to_convert = [
    'OpenTelemetry Support',
    'Distributed Tracing',
    'Ease of Use',
    'Real User Monitoring (RUM)',
    'Log Management',
    'Mobile Monitoring'
]

data[cols_to_convert] = data[cols_to_convert].astype(bool)
```

```python
data.to_csv('DATA.csv', index=False)
```
